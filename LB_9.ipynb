```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Дані для варіанта 2
x = np.array([1, 1.3, 1.7, 2.2, 2.8])
y = np.array([2.95, 3.89, 1.54, 3.38, 2.33])

# Побудова кубічного сплайну за допомогою CubicSpline
cs = CubicSpline(x, y)

# Генерація нових точок для гладкого графіку сплайна
x_new = np.linspace(min(x), max(x), 100)
y_new = cs(x_new)

# Побудова графіків
plt.figure(figsize=(8, 6))

# Графік кубічного сплайну
plt.plot(x_new, y_new, label="Кубічний сплайн", color='blue')

# Точки оригінальних даних
plt.scatter(x, y, label="Точки даних", color='red')

# Додаткове форматування графіка
plt.xlabel('x')
plt.ylabel('y')
plt.title('Кубічна інтерполяція сплайнами (Варіант 2)')
plt.legend()
plt.grid(True)

# Відображення графіка
plt.show()

# Виведення коефіцієнтів сплайну для кожного інтервалу
for i in range(len(x) - 1):
    coeffs = cs.c[:, i]  # Коефіцієнти для i-го інтервалу
    spline = f"S_{i+1}(x) = {coeffs[0]:.4f}*(x - {x[i]})^3 + {coeffs[1]:.4f}*(x - {x[i]})^2 + {coeffs[2]:.4f}*(x - {x[i]}) + {coeffs[3]:.4f}"
    print(spline)
```
