```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Функція для 2а
def ode_2a(x, y):
    return x + np.cos(y / 3)

# Функція для 2б
def ode_2b(x, y):
    return x + np.sin(y / np.sqrt(5))

# Параметри для задачі 2а
x0_a = 1.6
y0_a = 4.6
x_range_a = [1.6, 2.6]  # Інтервал для інтегрування

# Параметри для задачі 2б
x0_b = 1.8
y0_b = 2.6
x_range_b = [1.8, 2.8]  # Інтервал для інтегрування

# Розв'язок для задачі 2а
sol_a = solve_ivp(ode_2a, x_range_a, [y0_a], t_eval=np.linspace(x0_a, x_range_a[1], 100))

# Розв'язок для задачі 2б
sol_b = solve_ivp(ode_2b, x_range_b, [y0_b], t_eval=np.linspace(x0_b, x_range_b[1], 100))

# Виведення результатів
plt.figure(figsize=(10, 5))

# Графік для 2а
plt.subplot(1, 2, 1)
plt.plot(sol_a.t, sol_a.y[0], label="y' = x + cos(y/3)")
plt.title("Задача 2а")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Графік для 2б
plt.subplot(1, 2, 2)
plt.plot(sol_b.t, sol_b.y[0], label="y' = x + sin(y/√5)")
plt.title("Задача 2б")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Показати графік
plt.tight_layout()
plt.show()
```

```python
from scipy.integrate import odeint

def model_2a(y, x):
    return x + np.cos(y / 3)

def model_2b(y, x):
    return x + np.sin(y / np.sqrt(5))

x_a = np.linspace(1.6, 2.6, 100)
x_b = np.linspace(1.8, 2.8, 100)

# Рішення
y_a = odeint(model_2a, 4.6, x_a)
y_b = odeint(model_2b, 2.6, x_b)

# Візуалізація
plt.plot(x_a, y_a, label="Задача 2а")
plt.plot(x_b, y_b, label="Задача 2б")
plt.legend()
plt.show()
```
