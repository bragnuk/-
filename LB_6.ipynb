```python
import numpy as np  
import matplotlib.pyplot as plt 

# Введені дані для 2-го варіанту
x = np.array([-1, 0, 2, 3], dtype=float)
y = np.array([-16, -7, -1, 20], dtype=float)

# Точки, в яких потрібно знайти наближені значення
x_test = [-0.5, 0.5, 1.5, 2.5]

def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)

    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j])/(x[i] - x[j])
        p[i] = p_i

    return np.dot(y, p)

# Обчислення значень функції у заданих точках
results = {}
for xt in x_test:
    f_xt = lagrange_interpolation(x, y, xt)
    results[xt] = f_xt.round(4)
    print(f"Значення функції у точці {xt} = {f_xt.round(4)}")

# Побудова графіка
xnew = np.linspace(np.min(x), np.max(x), 100)
ynew = [lagrange_interpolation(x, y, i) for i in xnew]

plt.plot(x, y, 'o', xnew, ynew)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Polynomial (Варіант 2)')
plt.grid(True)
plt.show()
```
